```python
import numpy as np
from qiskit.visualization import plot_histogram
from Nature import *
```

```python
n_qubits = 10 
random_qc = QuantumCircuit(n_qubits)

# Create the circuit and add y-rotations for more fun
Create_random_circuit(random_qc,load_rots=False)
Show_Circuit(random_qc)
```

```python
# Choosing qubits. mode="two_player" for two players 
Lists = Get_qubits(n_qubits,mode="single_player")
```

```python
# Initializing the qubits selected
Circuit = Player_circuits(Lists,n_qubits,mode="single_player")
```

```python
# Printing the circuit (Spoiler alert!)
load_rotations(Circuit)
qc = Circuit.compose(random_qc)
qc.measure_all()    
qc.draw('mpl',style={'name':'iqx'})
```

```python
# Results (Check the largest bar)
counts = Run_circuit(qc)
plot_histogram(counts,figsize=(15,10))
```

```python
# The winner is...
alive,dead = Number_of_alive_cells(counts)
print("Live cells:",alive)
print("Dead cells:",dead)

who_wins(alive,dead)
```
